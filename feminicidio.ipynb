```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fe = pd.read_csv("feminicidio.csv",sep=";", encoding='ISO-8859-1')

fe.value_counts(sort=False)
```

```python
fe.columns.values
```

```python
vit = fe['vitimas']
ano = fe['ano_do_fato']
```

```python

plt.bar(vit, ano)
plt.title('Tabela_feminicideo')
plt.xlabel("vit", color='r')
plt.ylabel("ano", color='g')

plt.show()
```

```python
vit2 = [x for x in vit if x > 2 ]
print(vit2)

ano2 = [y for y in ano if y >= 2018]
ano3 = sorted(set(ano2))
print(ano3)
```

```python
plt.plot(vit2, ano3)
plt.xlabel('vit2', color='#00BFFF')
plt.ylabel('ano2', color='g')
plt.show()
```

```python
select = ['vitimas','ano_do_fato']
grafico = fe.filter(items=select)

grafico.value_counts(ascending=True)
```
